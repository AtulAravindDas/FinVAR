```python
import pandas as pd
financial_ratios['year'] = pd.to_datetime(financial_ratios['public_date']).dt.year
```

```python
df = df.rename(columns={'tic': 'TICKER'})
```

```python
import numpy as np

financial_ratios['epspx'] = np.nan


for index, row in financial_ratios.iterrows():
    ticker = row['TICKER']
    year = row['year']

    matching_rows = df[(df['TICKER'] == ticker) & (df['fyear'] == year)]

    if not matching_rows.empty:
        financial_ratios.loc[index, 'epspx'] = matching_rows['epspx'].iloc[0]
```
