```python
import numpy as np

financial_ratios['epspx'] = np.nan


for index, row in financial_ratios.iterrows():
    ticker = row['TICKER']
    year = row['year']

    matching_rows = new_df[(new_df['TICKER'] == ticker) & (new_df['fyear'] == year)]

    if not matching_rows.empty:
        financial_ratios.loc[index, 'epspx'] = matching_rows['epspx'].iloc[0]
```

## **Adding the revenues of the companies based on Ticker Names and Years and also generating the final dataframe to be used for Analysis**

```python
revenues_df=pd.read_csv('/content/drive/MyDrive/company_revenues.csv')
revenues_df.head()
```

```python
revenues_df.rename(columns={'tic':'TICKER','fyear':'year'},inplace=True)
```

```python
revenues_df.head()
```

```python
revenues_df.drop(columns=['datadate','gvkey','indfmt','consol','popsrc','datafmt','curcd','costat'],axis=1,inplace=True)
revenues_df.head()
```

```python
financial_ratios.head()
```

```python
financial_ratios=financial_ratios.merge(revenues_df[['TICKER','year','revt']],on=['TICKER','year'],how='left')
financial_ratios.head()
```

```python
epspx=financial_ratios['epspx']
epspx
```

```python
financial_ratios.drop(columns=['epspx'],axis=1,inplace=True)
financial_ratios['EPS']=epspx
financial_ratios.head()
```

```python
financial_ratios.isnull().sum()
```

```python
col='revt'
financial_ratios[col].fillna(financial_ratios[col].median(), inplace=True)
financial_ratios.isnull().sum()
```

```python
financial_ratios.rename(columns={'revt':'Revenue'},inplace=True)
```

```python
financial_ratios.to_csv('Financial_ratios_with_eps_and_revenue.csv')
```
